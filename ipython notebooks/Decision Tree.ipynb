# Decision Tree Regression

Decision Trees and their variants (random forests, extremely randomized trees) are extremely popular methods in modern machine learning, due to their ability to non-parametrically incorporate non-linearity, their white box nature, and their ease of implementation. 

## Pulling Data and Imports 

```python
import pandas as pd
import numpy as np
from sklearn import tree
import warnings
warnings.filterwarnings('ignore') ###Because of mse warning, fix at some point
%pylab inline
```

```python
x_data = pd.read_csv('C:\\Users\\smcdo\\OneDrive\\Documents\\Model_Framework\\Benchmarks\\x_data.csv', index_col=0)
y_data = pd.read_csv('C:\\Users\\smcdo\\OneDrive\\Documents\\Model_Framework\\Benchmarks\\y_data.csv', index_col=0, squeeze=True)
```

## Sklearn Implementation

```python
model = tree.DecisionTreeRegressor()
model = model.fit(x_data, y_data)
pred = model.predict(x_data)
print('Fitted Values: %s' % pred)
```
